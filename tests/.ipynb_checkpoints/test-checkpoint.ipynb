```python
import numpy as np
import matplotlib.pylab as plt

import sys
sys.path.append('..')
import kliep
```

# Generate Data

```python
dt = 1
t = np.arange(0,500,dt)
y = np.random.randn(len(t),3)*3
y_1 = np.array([[(i>100)*30] for i in range(len(t))])
y_2 = np.array([[(i>200)*30] for i in range(len(t))])
y_3 = np.array([[(i>300)*30] for i in range(len(t))])
y_4 = np.array([[(i>400)*30] for i in range(len(t))])
y_test = y+y_1+y_2+y_3+y_4
```

```python
plt.plot(y_test)
```

# Start Change Detection

```python
n_rf = 10
n_te = 20
k = 5
eta = 0.1
lam = 0.1
mu = 0.0001
```

```python
y_origin = y_test[:n_rf+n_te+k]
y_to_add = y_test[n_rf+n_te+k:]
```

```python
ddre = kliep.DensityRatioEstimator(y_origin,n_rf,n_te,k,eta,lam,mu)
```

```python
# Sequently use ddre.__call__() 
for y_new in y_to_add:
    ddre(y_new)
```

```python
ddre.plot_likelihood_ratio()
```
